```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


titanic = pd.read_csv('https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv')

df = titanic

df.head(10)
```

```python
size = [df.sex.value_counts()['male'],df.sex.value_counts()['female']]

labels = 'Male', 'Female'

plt.pie(size, labels = labels, autopct = '%1.1f%%', startangle = 90)
plt.axis('equal')

plt.show()
```

```python
f = df['fare']
a = df['age']
gender = df['sex']

colors = np.where(df['sex'] == 'female','r','g')

scatter_plt= plt.scatter(f,a, alpha=1,c=colors, s=10)
plt.title('Fare Paid vs Age')
plt.xlabel('Fare paid ')
plt.ylabel('Age')
plt.show()
```
